```python
import json
with open("we.json", "r") as f:
    we = json.load(f)
```

```python
with open("data.json","r") as f:
    data= json.load(f)
```

```python
counties = we["objects"]["counties"]
```

```python
states=we["objects"]["states"]
```

```python
data[4]
```

```python
for geo in counties["geometries"]:
    myid = geo["id"]
    for x in data:
        if int(str(x["state"])+str(x["county"]))==int(myid):
            geo["properties"]={"name": x["name"]}
            break
```

```python
for geo in states["geometries"]:
    myid = geo["id"]
    for x in data:
        if int(x["state"])==int(myid) and int(x["county"])==0:
            geo["properties"]={"name": x["name"]}
            break
```
